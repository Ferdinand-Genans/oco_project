#  <a name="C1">2 - Gradient Descent</a>

```python
#We will use this function to calculate the accuracy of all our models
def accuracy(a, b, memory_x, step=1):
  '''
  Parameters:
  a: test data (Matrix of size n*785)
  b: labels (Vector of size n)
  memory_x: List x_t, t=1,2,...,num_iterations
  Returns:
  accuracies: list of accuracies 
  '''
  accuracies = []
  for i in range(0, len(memory_x), step):
    x = memory_x[i]
    # Make predictions using the model
    y_pred = np.sign(a.dot(x))
    # Calculate the number of correct predictions
    false = np.sum(y_pred !=  b)
    # Calculate the total number of predictions
    total = len(b)
    # Calculate the accuracy as a proportion
    acc = false / total
    # Append the accuracy to the list
    accuracies.append(acc)
  return accuracies
```

## <a name="C1">Unconstrained GD</a>

```python
def gradient_descent_svm(a, b, learning_rate = 0.001, sqrt = False, num_iterations = 100):
  '''
  Parameters:
  a: training data (Matrix of size n*785)
  b: labels (Vector of size n)
  num_iterations: number of iterations
  Returns:  
  memory_x: List x_t, t=1,2,...,num_iterations
  '''
  n = a.shape[0] # number of samples
  # Initialize the parameters
  x_t = np.zeros(a.shape[1])
  memory_x = [x_t]
  # Set the number of iterations
  for t in range(1, num_iterations):
    # Calculate the gradient of the loss function
    if sqrt == True: 
        learning_rate = 1/np.sqrt(t)
    df = gradient_svm(x_t, a, b)
    # Update x_t
    x_t = x_t - learning_rate * df  
    memory_x.append(x_t)
  return memory_x


def gradient_svm(x_t, a, b, reg=0):
  '''
  Parameters: 
  x_t: current value of the parameters (Vector of size n)
  a: training data (Matrix of size n*785)
  b: labels (Vector of size n)
  reg: regularization parameter
  Returns:
  df: gradient of the loss function
  '''
  n = a.shape[0] # number of samples
  df = np.zeros(a.shape[1])
  # Sum of the gradients of hinge loss functions
  for i in range(n):
    if b[i] * np.dot(a[i], x_t) <= 1:
      df += -b[i] * a[i]
  df /= n #divide by the number of hinge losses
  # Gradient of the regularization term
  df += reg*x_t
  return df
```

```python
#Illustration of GD with various learning rate
learning_rate = [0.1, 0.01, 0.001]
for lr in learning_rate: 
    memory_x = gradient_descent_svm(a, b, lr, num_iterations=100)
    accuracy_tab = accuracy(a_test, b_test, memory_x)
    x = np.arange(0, len(memory_x), 1)
    min_str = str(min(accuracy_tab).round(4))
    label = "lr = " + str(lr) + ", min = " + min_str
    plt.loglog(x, accuracy_tab, label = label)
memory_x = gradient_descent_svm(a, b, 1, sqrt=True, num_iterations=100)
accuracy_tab = accuracy(a_test, b_test, memory_x)
x = np.arange(0, len(memory_x), 1)
min_str = str(min(accuracy_tab).round(4))
label = "lr = 1/sqrt(t), min = " + min_str
plt.loglog(x, accuracy_tab, label = label)
plt.xlabel('number of iterations')
plt.ylabel('Accuracy')
plt.ylim(0.01, 0.11)
plt.title('Accuracy Gradient descent, varying the learning rate')
plt.legend()
plt.show()
```

## <a name="C1">Projected GD</a>

```python
def projection_simplex(x):
    '''
    Parameters:
    x: Vector of size n
    Returns:
    P_x: projection of x on the probability simplex
    '''
    # Check if x is already a probability vector
    if np.sum(x) == 1 and np.alltrue(x >= 0):
        P_x = x
    else: 
        # Sort the elements of x in decreasing order
        x_sorted = sorted(x, reverse=True)
        # Calculate the cumulative sum of the sorted elements
        cumsum = np.cumsum(x_sorted)
        # Find the largest index such that v_sorted[i] - (cumsum[i] - 1) / (i+1) > 0
        d_0 = np.nonzero(x_sorted - (cumsum - 1) / (np.arange(len(x)) + 1) > 0)[0][-1]
        # Calculate the projection
        theta = (cumsum[d_0] - 1) / (d_0 + 1)
        # Project v onto the probability simplex
        P_x = np.maximum(x - theta, 0) #SoftTreshold(x, theta)
    return P_x 

def projection_l1_ball(x, z):
    '''
    Parameters:
    x: Vector of size n
    z: radius of the l1 ball
    Returns:
    P_x: projection of x on the l1 ball of radius z
    '''
    #Check if x is already in the l1 ball of radius z
    if np.linalg.norm(x, ord=1) <= z:
        P_x = x
    else: 
        sign_x = np.sign(x)
        abs_x = np.abs(x)
        P_x_simplex = projection_simplex(abs_x / z)
        P_x = sign_x * P_x_simplex * z
    return P_x

def ProjectedGD_svm(a, b, reg, z, num_iterations=100):
    '''
    Parameters:
    a: training data (Matrix of size n*785)
    b: labels (Vector of size n)
    reg: regularization parameter
    z: radius of the l1 ball
    num_iterations: number of iterations
    Returns:  
    x_t: our minimizer
    '''
    n = a.shape[0] # number of samples
    # Initialize the parameters
    x_t = np.zeros(a.shape[1])
    memory_x = [x_t]
    # Set the number of iterations
    for t in range(1, num_iterations):
        # Calculate the gradient of the loss function
        df = gradient_svm(x_t, a, b, reg)
        # Update x_t
        learning_rate = 1/(reg*t)
        y_t = x_t - learning_rate * df  
        # Project x_t on the l1 ball of radius z
        x_t = projection_l1_ball(y_t, z)
        memory_x.append(x_t)
    return memory_x
```

```python
def SGD_svm(a, b, learning_rate = 0.001, reg= 0, num_iterations = 10000):
  '''
  Parameters:
  a: training data (Matrix of size n*785)
  b: labels (Vector of size n)
  reg: regularization term
  num_iterations: number of iterations
  Returns:  
  memory_x: List x_t, t=1,2,...,num_iterations
  '''
  n = a.shape[0] # number of samples
  # Initialize the parameters
  x_t = np.zeros(a.shape[1])
  memory_x = [x_t]
  # Array of random indices 
  rand_indices = np.random.choice(n-1, size=num_iterations, replace=True)
  for t in range(1, num_iterations):
    index = rand_indices[t]
    # Calculate the gradient of the loss function+ REG and update x_t
    if b[index] * np.dot(a[index], x_t) <= 1:
      x_t = x_t -learning_rate*(-b[index] * a[index] + reg*x_t)
    else:
      x_t = x_t -learning_rate*reg*x_t #gradient of the regularisation term
    #append the new x_t to the list
    memory_x.append(x_t)
  return memory_x

    #previous_average = memory_x[t-1] *t-1 #old average multiply by the number of previous iterations
    #new_average = (previous_average + x_t)/t #new average
    #memory_x.append(new_average)
    
def ProjectedSGD_svm(a, b, reg = 0.5, z=100, num_iterations = 100):
  '''
  Parameters:
  a: training data (Matrix of size n*785)
  b: labels (Vector of size n)
  reg: regularization term
  z: radius of the l1 ball
  num_iterations: number of iterations
  Returns:  
  memory_x: List x_t, t=1,2,...,num_iterations
  '''
  n = a.shape[0] # number of samples
  # Initialize the parameters
  x_t = np.zeros(a.shape[1])
  memory_x = [x_t]
  # Array of random indices 
  rand_indices = np.random.choice(n-1, size=num_iterations, replace=True)
  for t in range(1, num_iterations):
    learning_rate = 1/(reg*t)
    index = rand_indices[t]
    # Calculate the gradient of the loss function+ REG and update x_t
    if b[index] * np.dot(a[index], x_t) <= 1:
      y_t = x_t -learning_rate*(-b[index] * a[index] + reg*x_t)
    else:
      y_t = x_t -learning_rate*reg*x_t #gradient of the regularisation term
    x_t = projection_l1_ball(y_t, z)
    #append the new x_t to the list
    memory_x.append(x_t)
  return memory_x


def cumulative_mean(X):
  # Initialize the cumulative mean to the first vector
  cum_mean = X[0]
  # Initialize a list to store the cumulative means
  cum_means = [cum_mean]
  # Iterate through the remaining vectors
  for i in range(1, len(X)):
      # Update the cumulative mean
      cum_mean = (i * cum_mean + X[i]) / (i + 1)
      # Append the cumulative mean to the list
      cum_means.append(cum_mean)
  
  return cum_means
```

## <a name="C1">Stochastic Mirror Descent</a>

```python
def SOMD_svm(a, b, z=100, num_iterations = 100):
  '''
  Parameters:
  a: training data (Matrix of size n*785)
  b: labels (Vector of size n)
  reg: regularization term
  z: radius of the l1 ball
  num_iterations: number of iterations
  Returns:  
  memory_x: List x_t, t=1,2,...,num_iterations
  '''
  n = a.shape[0] # number of samples
  # Initialize the parameters
  x_t = np.zeros(a.shape[1])
  y_t = np.zeros(a.shape[1])
  memory_x = [x_t]
  # Array of random indices 
  rand_indices = np.random.choice(n-1, size=num_iterations, replace=True)
  for t in range(1, num_iterations):
    learning_rate = 1/np.sqrt(t)
    index = rand_indices[t]
    # Calculate the gradient of the loss function and update x_t
    if b[index] * np.dot(a[index], x_t) <= 1:
      y_t = y_t -learning_rate*(-b[index] * a[index])
    x_t = projection_l1_ball(y_t, z)
    #append the new x_t to the list
    memory_x.append(x_t)
  return memory_x
```

## <a name="C1">Stochastic Exponentiated Gradient</a>

```python
def SEG_svm(a, b, z=100, lr=1, num_iterations = 100):
    '''
    Parameters:
    a: training data (Matrix of size n*785)
    b: labels (Vector of size n)
    reg: regularization term
    z: radius of the l1 ball
    num_iterations: number of iterations
    Returns:  
    memory_x: List x_t, t=1,2,...,num_iterations
    '''
    n = a.shape[0] # number of samples
    n, d = a.shape
    # Initialize the parameters
    x_t = np.zeros(a.shape[1])
    memory_x = [x_t]
    weights = np.ones(d)/(2*d)
    tetap = np.zeros(d)
    tetam = np.zeros(d)
    # Array of random indices 
    rand_indices = np.random.choice(n-1, size=num_iterations, replace=True)
    for t in range(1, num_iterations):
        learning_rate = lr*(1/np.sqrt(t))
        index = rand_indices[t]
        # Calculate the gradient of the loss function + reg
        grad = np.zeros(d)
        if b[index] * np.dot(a[index], x_t) <= 1:
            grad = -b[index] * a[index]
        tetam = tetam - learning_rate*grad
        tetap = tetap + learning_rate*grad
        tetat = np.r_[tetam, tetap]
        # Update the weights
        weights = np.exp(tetat)/np.sum(np.exp(tetat))
        # Update x_t
        x_t = z * (weights[0:d] - weights[d:])
        #append the new x_t to the list
        memory_x.append(x_t)
    return memory_x
```

## <a name="C1">Stochastic AdaGrad</a>

```python
def weighted_simplex(x, D):
    '''
    x: vector
    D: matrix of weights
    returns: projection of x on the weighted simplex
    '''

    D_x = np.abs(np.dot(D, x))
    sorted_indices = np.argsort(-D_x, kind='quicksort')
    sum_x = np.cumsum(x[sorted_indices])
    sum_d = np.cumsum(1 / np.diag(D)[sorted_indices])
    res = D_x[sorted_indices] - (sum_x - 1) / sum_d
    d0 = np.max(np.where(res > 0)[0])
    theta = (sum_x[d0] - 1) / sum_d[d0]

    return np.dot(np.linalg.inv(D), np.maximum(0, D_x - theta))
    
def weighted_projection_l1(x, d, z=1):
    '''
    Weighted projection on the l1-ball of
    x: input vector
    d: weights
    z: radius of the l1 ball
    returns: weighted projection of x on the l1-ball
    '''
    if np.sum(np.abs(x)) <= z:
        return x
    else:
        proj = weighted_simplex(abs(x) / z, np.diag(d))
    return z * np.sign(x) * proj

def Adagrad_svm(a, b, z=100, reg=0.01, num_iterations = 100):
    '''
    Parameters:
    a: training data (Matrix of size n*785)
    b: labels (Vector of size n)
    reg: regularization term
    z: radius of the l1 ball
    num_iterations: number of iterations
    Returns:  
    memory_x: List x_t, t=1,2,...,num_iterations
    '''
    n, d = a.shape
    # Initialize the parameters
    x_t = np.zeros(a.shape[1])
    memory_x = [x_t]
    St = np.ones(d)/1e-6
    # Array of random indices 
    rand_indices = np.random.choice(n-1, size=num_iterations, replace=True)
    for t in range(1, num_iterations):
        index = rand_indices[t]
        # Calculate the gradient of the loss function + reg
        grad = reg*x_t
        if b[index] * np.dot(a[index], x_t) <= 1:
            grad += -b[index] * a[index]
        St = St + grad**2
        Dt = np.sqrt(St)
        y_t = x_t -  grad/(Dt + 1e-6)
        x_t = weighted_projection_l1(y_t, np.diag(Dt), z)
        #append the new x_t to the list
        memory_x.append(x_t)
    return memory_x
```

## <a name="C1">Stochastic ONS</a>

```python
def ONS_svm(a, b, z, reg, gamma, num_iterations):
    # Initialization 
    n, d = a.shape
    x_t = np.zeros(a.shape[1])
    memory_x = [np.zeros(a.shape[1])]
    A_t = (1/gamma**2)*np.identity(d)
    # Sample iid from 0 to n - 1
    I_t = np.random.randint(0, n, size=num_iterations)
    for t in range(num_iterations-1):
        step = 1/gamma
        i = random.randint(0, n-1)
        grad_t = reg*x_t
        if b[i] * np.dot(a[i], x_t) <= 1:
            grad_t += -b[i] * a[i]
        grad_matrix = np.outer(grad_t, grad_t) # gradients matrix (hessian approximation)
        A_t+= grad_matrix
        A_inv_t = np.linalg.inv(A_t)
        y_t = x_t - step*np.dot(A_inv_t, grad_t)
        x_t = weighted_projection_l1(y_t, np.diag(A_t), z)
        memory_x.append(x_t)
    return memory_x
```

### SVRG Algorithm 


```python
def SVRG_svm(a, b, option =1, learning_rate = 0.001, reg= 0.05, m=100, num_iterations = 10000):
  '''
  Parameters:
  a: training data (Matrix of size n*785)
  b: labels (Vector of size n)
  reg: regularization term
  num_iterations: number of iterations
  Returns:  
  memory_x: List x_t, t=1,2,...,num_iterations
  '''
  n = a.shape[0] # number of samples
  # Initialize the parameters
  x_t = np.zeros(a.shape[1])
  w = np.zeros(a.shape[1])
  memory_x = [x_t]
  #compute the full gradient
  grad_f = np.zeros(a.shape[1])
  # Array of random indices 
  rand_indices = np.random.choice(n-1, size=num_iterations, replace=True)
  if option == 1:
    for t in range(1, num_iterations):
      if t % m == 0 and t != 0:
        grad_f = gradient_svm(x_t, a, b, reg)
        w = x_t
      # Pick a random index
      index = rand_indices[t]
      # Estimate of the gradient with reduced variance+ REG and update x_t
      grad_f_t = grad_f +reg*x_t #part not depending on x_t and the random index
      #Hinge loss gradient x_t, random index
      if b[index] * np.dot(a[index], x_t) <= 1:
        grad_f_t += -b[index] * a[index]
      #Hinge loss gradient w, random index
      if b[index] * np.dot(a[index], w) <= 1:
        grad_f_t += -b[index] * a[index]
      x_t = x_t - learning_rate * grad_f_t #update x_t
      #append the new x_t to the list
      memory_x.append(x_t)
      #update the full gradient and w every m iterations
  elif option == 2:
    w = np.zeros(a.shape[1])
    w_candidates = np.zeros((m, a.shape[1]))
    for t in range(1, num_iterations):
      if t % m == 0 and t != 0:
        index_w = np.random.randint(0, m)
        w = w_candidates[index_w]
        grad_f = gradient_svm(w, a, b, reg)
      # Pick a random index
      index = rand_indices[t]
      # Estimate of the gradient with reduced variance+ REG and update x_t
      grad_f_t = grad_f +reg*x_t #part not depending on x_t and the random index
      #Hinge loss gradient x_t, random index
      if b[index] * np.dot(a[index], x_t) <= 1:
        grad_f_t += -b[index] * a[index]
      #Hinge loss gradient w, random index
      if b[index] * np.dot(a[index], w) <= 1:
        grad_f_t += -b[index] * a[index]
      x_t = x_t - learning_rate * grad_f_t #update x_t
      #append the new x_t to the list
      memory_x.append(x_t)
      #update the full gradient and w every m iterations
      w_candidates[t % m] = x_t
      #update the full gradient every m iterations
  return memory_x
```
